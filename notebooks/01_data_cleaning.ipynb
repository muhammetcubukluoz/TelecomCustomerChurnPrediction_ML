```python
data.columns = data.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
for col in data:
    print(f"{col} \n", data[col].unique())
    print()
```

```python
data.drop(columns=["customerid", "count", "country", "state", "lat_long"], inplace=True)
```

```python
cols_to_replace = ["multiple_lines", "online_security", "online_backup",
                   "device_protection", "tech_support", "streaming_tv", "streaming_movies"]

for col in cols_to_replace:
    data[col] = data[col].replace({"No internet service": "No", "No phone service": "No"})
```

```python
for cols in ["multiple_lines", "online_security", "online_backup",
                   "device_protection", "tech_support", "streaming_tv", "streaming_movies"]:
    print(f"{cols} \n", data[cols].unique())
    print()
```

```python
data.drop(columns=["churn_label"], inplace=True)
```

```python
data["total_charges"] = pd.to_numeric(data["total_charges"], errors="coerce")
data["total_charges"].isnull().sum()
```

```python
data[np.isnan(data["total_charges"])]
```

```python
data.fillna(data["total_charges"].mean())
```
